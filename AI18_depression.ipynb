```python
# drop column 'days' and 'madrs2'
# 'days' meant how many days the patient was treated which we do not need in this project.
# 'madrs2' is the new madrs score after the treatment. We need only one before treatment.
df.drop(['days','madrs2'], axis=1, inplace=True)
```

```python
# show the new statistics summary of the file
df.describe()
```

```python
# slied csv row 23 to 54 due to too many missing values on the bottom parts of csv file
df.drop(labels=range(23, 55), axis=0)
```

```python
# change column 'gender' from integer to object
df['gender'] = df['gender'].replace({1 : 'Female'})
df['gender'] = df['gender'].replace({2 : 'Male'})
```

```python
# check if the change was succesful 
df.head()
```

```python
# create a pie chart on the percentage of depression per gender
Gender = df['gender']
Gender.value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Percent of gender with depression')
```

```python
# create a bar chart on MADRS score by gender to see which gender has higher score
plt.bar('gender', 'madrs1', data = df, linewidth = 1)
plt.xlabel('Gender')
plt.xticks(rotation = 90)
plt.ylabel('MADRS score')
plt.title('MADRS score by gender')
plt.show()
```

```python
# create a pie chart on the percentage of depression per age group
Age = df['age']
Age.value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Percent of age group with depression')
```

```python
# create a bar chart on MADRS score by age group to see which group has higher score
plt.bar('age', 'madrs1', data = df, linewidth = 1)
plt.xlabel('Age group')
plt.xticks(rotation = 90)
plt.ylabel('MADRS score')
plt.title('MADRS score by age group')
plt.show()
```

```python
# check the column types again for the pie chart
df.info()
```

```python
# change 'afftype' from integer to object
df['afftype'] = df['afftype'].replace({1.0 : 'Biopolar II'})
df['afftype'] = df['afftype'].replace({2.0 : 'Unipolar Depressive'})
df['afftype'] = df['afftype'].replace({3.0 : 'Biopolar I'})
```

```python
# create a pie chart on the percentage of depression per afftype
Afftype = df['afftype']
Afftype.value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Percent of afftype with depression')
```

```python
# create a bar chart on MADRS score by afftype to see which type has higher score
df['madrs1'] = df['madrs1'].astype(str) 
df['afftype'] = df['afftype'].astype(str) 
plt.bar('afftype', 'madrs1', data = df, linewidth = 1)
plt.xlabel('Afftype')
plt.xticks(rotation = 90)
plt.ylabel('MADRS score')
plt.title('MADRS score by afftype')
plt.show()
```
